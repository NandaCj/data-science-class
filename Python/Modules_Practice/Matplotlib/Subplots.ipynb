```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)
```

```python
fig, axes = plt.subplots(2, 2,)
axes[0, 1].plot(x, y)
axes[1, 1].scatter(x, y)
```
