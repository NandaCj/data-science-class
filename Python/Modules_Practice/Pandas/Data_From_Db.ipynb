```python
import pandas as pd
import psycopg2
%matplotlib inline
```

```python
connection = psycopg2.connect(user = "postgres",
                                  password = "postgres",
                                  host = "127.0.0.1",
                                  port = "5432",
                                  database = "stock")
cursor = connection.cursor()
```

```python
result = cursor.execute("select * from sector")
print(result)
```

```python
df = pd.read_sql('select * from sector', connection)
df['sector'] = df['sector'].str.rstrip("\n")
df.head()
```

```python
df_g = df.groupby('industry').count()
```

```python
df_g.reset_index(inplace=True)
# dfg = dfg.sort_values()
```

```python
df_g.plot(x='industry', y='sector', kind='bar')
```
