```python
%matplotlib inline 
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud as WC
from wordcloud import STOPWORDS 
```

# Loading Dataset and First Look

```python
# Setting Path
Curr_Dir = os.path.dirname(os.getcwd())
data_path = os.path.join(Curr_Dir , 'Datasets\Author_Identification')
sample = os.path.join(data_path, 'sample_submission.csv')
train = os.path.join(data_path, 'train.csv')
test = os.path.join(data_path, 'test.csv')

# Loading Data
Sample_Data = pd.read_csv(sample)
Train_Data = pd.read_csv(train)
Test_Data = pd.read_csv(test)

# Shapes

print(Sample_Data.shape)
print(Train_Data.shape)
print(Test_Data.shape)
```

# Using Test_Data to make Word Cloud

```python
Text = Test_Data.text[:1].str.cat(sep=' ')
print(Text)
wc = WC().generate(Text)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
STOPWORDS.add('Still')
Text = Test_Data.text[:1].str.cat(sep=' ')
wc = WC(stopwords=STOPWORDS).generate(Text)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
print(len(STOPWORDS))
STOPWORDS.add('upon')
```

```python
Text = Test_Data.text[:100].str.cat(sep=' ')
wc = WC(background_color ='black', stopwords=STOPWORDS)
wc.generate(Text)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
