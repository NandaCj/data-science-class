# Notes:
    - As with all the transformations, it is important to fit the scalers to the training data only, not to the full dataset (including the test set). Only then can you use them to transform the training set and the test set (and new data).

```python
import os
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
breast_cancer_dataset = 'C:\\Users\\nandpara\\PycharmProjects\\Modules_Overview1\\Dataset_Analysis\\breast_cancer_dataset.csv'
df = pd.read_csv(breast_cancer_dataset)
```

# Removing non numeric columns

```python
non_numeric_columns = ['id', 'Unnamed: 32', 'diagnosis']
df.drop(non_numeric_columns, axis=1, inplace=True)
```

```python
df.head()
```

### sklearn.preprocessing.StandardScaler returns numpy arrary for fit_transform of Pandas dataframe

```python
scaled = scaler.transform(df)
print(scaled.shape)
print(scaled)
```

### So do as below to get the scaled data back on Dataframe
- u can also standardize specific columns as well

```python
df[df.columns] = scaler.fit_transform(df[df.columns])
```
