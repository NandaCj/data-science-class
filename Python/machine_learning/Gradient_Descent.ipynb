# Gradient Descent 

    - GD is used to to reduce the computational complexity of Normal Equations
# Types of Gradient Descent:
    - Batch Gradient Descent - use complete batch dataset
    - Stochastic Gradient Descent - use 1 dataset for every iteration
    - Mini-Batch Gradient Descent - use small batch dataset


```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Forming Data with function y = mx (here m = 2)
m = 5
N = 10
x = np.arange(1, N+1)
y = m * x

print(x)
print(y)
```

```python
plt.plot(x, y)
```

```python
# Cost Function :

theta = 2# random Initial Theata Value (choose any value)
hypothesis = np.dot(theta, x)
hypothesis
```

```python
plt.plot(x, y, label = 'Actual theta = {}'.format(m))
plt.plot(x, hypothesis, label = 'Predicted with thetha = {}'.format(theta))
plt.legend()
```

# Batch Gradient Descent 

    - Use complete dataset for calculating Modal Parameters(Theta)

```python
def BatchGradientDescent(x, y, theta, learning_rate, iterations, tolerance):
    plt.plot(x, y, color = 'black', label='Actual theta = {}'.format(m))
    theta_list = [theta]
    for i in range(iterations):
        
        hypothesis = np.dot(x, theta)
        error = hypothesis - y
        cost = np.sum(error ** 2) / 2*m
        gradient = np.dot(x.transpose(), error) / m 
        theta = theta - learning_rate * gradient 
        print(theta)
        theta_list.append(theta)
        plt.plot(x, hypothesis, label ='Predicted with thetha = {}'.format(theta) )
        
        if tolerance :
            if m - theta <= tolerance :
                break
    plt.legend()
    plt.rcParams["figure.figsize"] = (80,20)

    plt.show()
    
    return theta_list
```

```python
iterations = 30
tolerance = False
learning_rate = 0.001
initial_random_theta = 10
theta_list = BatchGradientDescent(x, y , theta=initial_random_theta, learning_rate=learning_rate, iterations=iterations , tolerance = tolerance)
```

```python
import matplotlib.pyplot as plt
print(iterations, len(theta_list))
plt.rcParams["figure.figsize"] = (10,10)
plt.scatter(range(1, iterations +2), theta_list, )
```

# Stochastic Gradient Descent
    - Randomize the Data
    - only one training example in every iteration to compute the gradient of cost function
    - one might not achieve accuracy, but the computation of results are faster

# Mini Batch Gradient Descent 
    - shuffle the observations
    - Divide them into batches 
    - Run each batch seperately like Batch GD
